```python
import pandas as pd
tables = pd.read_html("https://en.wikipedia.org/wiki/List_of_sovereign_states_and_dependent_territories_by_birth_rate")
countries = tables[0]
countries
```

```python
countries.columns
```

```python
countries.columns = ['country','prb2022','ciawf2023']
```

```python
display(countries.head(5))
```

All countries

Select all countries, but not the dependendent territories. Include all columns.

```python
countries[countries['country'].str.contains('\(')].sort_values(by='country')
```

PRB birthrate over 40

Select the countries that have a birthrate higher than 40 according to the Population Reference Bureau. Select only the columns with the country name and the PRB birth rate.

```python
#birth rate larger than 40
countries[countries['prb2022'] > 40][['country','prb2022']]
```

```python
stan_mask = countries.country.str.contains("stan")
prb2022_low_br = countries.prb2022 < 25
countries.loc[stan_mask & prb2022_low_br,["country", "prb2022"]]
```

```python
def dropping_birth_rate(country):
    drop = country.cia13 - country.cia20
    return drop > 7

countries.loc[lambda country: dropping_birth_rate(country), ["country", "cia13", "cia20"]]
```
